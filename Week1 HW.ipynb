## 2.計算日子至今過了多久

```python
from ipywidgets import interact
from ipywidgets import interact_manual
import time
```

```python
y=input("年(西元)")
m=input("月")
d=input("日")
date=time.strptime(f'{y}-{m}-{d} 00:00', '%Y-%m-%d %H:%M')
a=0
while a!=1: 
        now=time.time()
        interval=now-time.mktime(date)
        m, s = divmod(interval, 60)
        h, m = divmod(m, 60)
        d, h = divmod(h, 24)
        y, d = divmod(d, 365)
        k="{0}年{1}天{2}小時{3}分{4}秒".format(int(y),int(d),int(h),int(m),int(s))
        print(k)
        time.sleep(1)
```
