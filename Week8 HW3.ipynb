```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_tmp=pd.read_html("https://tw.stock.yahoo.com/s/list.php?c=%AD%B9%AB%7E&rr=0.96242800%201573461429",encoding="big5")
```

```python
df=df_tmp[2]
```

```python
df
```

### 刪除部必要的欄位

```python
df.drop("選擇", axis=1, inplace=True)
df.drop("漲跌", axis=1, inplace=True)
```

```python
df
```

### 增加欄位、找出相差值前五名

```python
df["相差"] = df.最高 - df.最低
```

```python
df.sort_values("相差",ascending=False).head(5)
```

### 成交價前五名

```python
df.sort_values("成交",ascending=False).head(5)
```

### 成交價折線圖

```python
df.成交.plot()
```

### 成交張數前五名

```python
df.sort_values("張數",ascending=False).head(5)
```

### 漲跌前五名

```python
df.insert(6,'漲跌',(df.昨收-df.成交))
```

```python
df.sort_values("漲跌",ascending=False).head(5)
```
