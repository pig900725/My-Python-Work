```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_tmp=pd.read_html("http://school.hwes.ylc.edu.tw/html/taiwan/4-21.htm")
```

```python
df=df_tmp[0]
```

```python
df
```

### 增加四季的雨量

```python
df.insert(7,'上半年',(df.一月+df.二月+df.三月+df.四月+df.五月+df.六月))
df.insert(14,'下半年',(df.七月+df.八月+df.九月+df.十月+df.十一月+df.十二月))
```

```python
df
```

### 上下半年的雨量排名

#### 上半年

```python
df_first=df[["地名","一月","二月","三月","四月","五月","六月","上半年"]]
```

```python
df_first.sort_values(by="上半年",ascending=False).head()
```

#### 下半年

```python
df_second=df[["地名","七月","八月","九月","十月","十一月","十二月","下半年"]]
```

```python
df_second.sort_values(by="下半年",ascending=False).head()
```

#### 全年雨量排名

```python
df_total=df[["地名","合計"]]
```

```python
df_total.sort_values(by="合計",ascending=False).head()
```
